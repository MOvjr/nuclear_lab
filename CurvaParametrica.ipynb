```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función paramétrica
def parametric_curve(t):
    x = np.cos(t)
    y = np.sin(t)
    return x, y

# Generar puntos para t en el rango deseado
t_values = np.linspace(0, 2*np.pi, 100)

# Calcular los valores de x y y para la curva paramétrica
x_values, y_values = parametric_curve(t_values)

# Graficar la curva paramétrica
plt.plot(x_values, y_values, label='Curva paramétrica')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Curva paramétrica: x = cos(t), y = sin(t)')
plt.axis('equal')  # Para que los ejes tengan la misma escala
plt.legend()
plt.grid(True)
plt.show()
```
